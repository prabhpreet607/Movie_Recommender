```python
import numpy as np
import pandas as pd
```

```python
credit=pd.read_csv("tmdb_5000_credits.csv")
movie=pd.read_csv("tmdb_5000_movies.csv")
```

```python
movie=movie.merge(credit,on="title")
```

```python
movie=movie[["genres","id","keywords","title","overview","cast","crew"]]
```

```python
movie.head()
```

```python
movie["genres"][0]
```

```python
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i["name"])
    return l    
        
        
```

```python
movie["genres"]= movie["genres"].apply(convert)
```

```python
movie.head()
```

```python
movie["keywords"][0]
```

```python
movie["keywords"]=movie["keywords"].apply(convert)
```

```python
movie.head(1)
```

```python
movie["cast"][0]
```

```python
def convert3(text):
    l=[]
    counter=0
    for i in ast.literal_eval(text):
            if counter<5:
                l.append(i["name"])
                counter+=1
    return l    
        
```

```python
movie["cast"]=movie["cast"].apply(convert3)
```

```python
movie.head()
```

```python
movie["crew"][0]
```

```python
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i[ "job"]== "Director":
            l.append(i["name"])
    return l        
```

```python
movie["crew"]=movie["crew"].apply(fetch_director)
```

```python
movie.head()
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie["cast"][0]
```

```python
def remove_spcae(text):
    l=[]
    for i in text:
        l.append(i.replace(" ",""))
    return l   
        
```

```python
movie["cast"]=movie["cast"].apply(remove_spcae)
movie["crew"]=movie["crew"].apply(remove_spcae)
movie["genres"]=movie["genres"].apply(remove_spcae)
movie["keywords"]=movie["keywords"].apply(remove_spcae)
```

```python
movie.head()
```

```python
movie["overview"]=movie["overview"].apply(lambda x:x.split())
```

```python
movie.head()
```

```python
movie["tag"]=movie["genres"]+movie["keywords"]+movie["overview"]+movie["cast"]+movie["crew"]
```

```python
movie.drop(columns=['overview','genres','keywords','cast','crew'],inplace=True)
```

```python
movie.head()
```

```python
movie["tag"]=movie["tag"].apply(lambda x:"  ".join(x))
```

```python
movie.head()
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(movie["tag"]).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
def recommender(new_movie):
    index=movie[movie["title"]==new_movie].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:6]:
        print(movie.iloc[i[0]].title)
        
```

```python
recommender("Batman")
```
